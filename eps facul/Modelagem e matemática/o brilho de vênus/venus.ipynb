```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
import plotly.express as px
```

```python
luminosidade = pd.DataFrame({'cos_theta':[],'r_theta':[]})
```

```python
def get_r_theta (cos_theta): # Os valores de R (distancia de Sol a VÃªnus) e d57 (distancio de Sot a 
  R  = 10.81*10**7 
  dST =  14.95*10**7 
  resultado =  np.sqrt(R**2+ dST**2 -2*dST*R*cos_theta) 
  return resultado
```

```python
luminosidade['cos_theta'] = np.arange(-1,1,0.01)
```

```python
luminosidade.head()
```

```python
luminosidade['r_theta'] = luminosidade['cos_theta'].apply(get_r_theta)
```

```python
luminosidade.head()
```

```python
def get_r(r):
  K = 1
  #Os valores de R (distancia de Sol a Venus) e dST (distancia de Sol a Terra) 
  dST = 14.95*10**7
  R = 10.81 *10**7
  resultado = (K/(4* R)) * (1/r**3) *(R**2 - dST**2 + r**2+2* R* r)
  return resultado
```

```python
luminosidade['r_k1'] =  luminosidade['r_theta'].apply(get_r)
```

```python
luminosidade.head()
```

```python
fig = px.line(luminosidade,x='r_theta',y='r_k1')
fig.show()
```
